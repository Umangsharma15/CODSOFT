```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
train_df = pd.read_csv('/content/fraudTrain.csv')

```

```python
test_df = pd.read_csv('/content/fraudTest.csv')
```

```python
df = pd.concat([train_df, test_df])
```

```python
drop_columns = ['cc_num', 'first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num', 'unix_time']
df = df.drop(columns=drop_columns, errors='ignore')
```

```python
train_size = len(train_df)
train_df, test_df = df.iloc[:train_size], df.iloc[train_size:]
```

```python
X_train, y_train = train_df.drop(columns=['is_fraud']), train_df['is_fraud']
X_test, y_test = test_df.drop(columns=['is_fraud']), test_df['is_fraud']
```

```python
print(y_train[:10])
print(y_train.dtype)
```

```python
y_train = np.round(y_train).astype(int)
y_test = np.round(y_test).astype(int)
```

```python
y_train = pd.cut(y_train, bins=3, labels=[0, 1, 2])
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier()
}
```

```python

train_df = train_df.dropna()
test_df = test_df.dropna()


train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')


X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)


target_imputer = SimpleImputer(strategy='most_frequent')

y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)



y_train = target_imputer.fit_transform(y_train)
y_test = target_imputer.transform(y_test)
y_train = y_train.ravel()
y_test = y_test.ravel()
```

```python
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    results[name] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, average='weighted'),
        'Recall': recall_score(y_test, y_pred, average='weighted'),
        'F1 Score': f1_score(y_test, y_pred, average='weighted')
    }
```

```python
results_df = pd.DataFrame(results).T
print(results_df)
```
